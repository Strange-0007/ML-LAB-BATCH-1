##Data Preprocessing

Aim of the Experiment.
The main aim of this experiment is to preprocess the given dataset. The database is created and is available in the file sample.csv.



Load the data from "sample.csv", apply Label Encoding, a scaling technique, and Binarization to the data, and display the results after each transformation.

## Data loading

Load the data from "sample.csv" into a dataframe.


```python
import pandas as pd

df = pd.read_csv('preprocessing.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df['gender_encoded'] = label_encoder.fit_transform(df['gender'])

display(df.head())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df['Marks_scaled'] = scaler.fit_transform(df[['Marks']])

display(df.head())
```

```python
from sklearn.preprocessing import Binarizer

binarizer = Binarizer(threshold=50)

df['Marks_binarized'] = binarizer.fit_transform(df[['Marks']])

display(df.head())
```
